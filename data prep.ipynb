```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
categorical_cols = insurance_df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    insurance_df[col] = le.fit_transform(insurance_df[col])
```

```python
insurance_df.head()
```

```python
plt.figure(figsize = (18, 12))

corr = insurance_df.corr()

sns.heatmap(data = corr, annot = True)
plt.show()
```

```python
insurance_df['claim sum']=insurance_df['vehicle_claim']+ insurance_df['property_claim'] + insurance_df['injury_claim']

insurance_df.head()
```

```python
insurance_df['claim to total ratio']=insurance_df['claim sum'] / insurance_df['total_claim_amount'].replace(0,pd.NA)
insurance_df.head()
```

```python
plt.Figure(figsize=(10,10))
sns.lineplot(insurance_df['claim to total ratio'])
```
